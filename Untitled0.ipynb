```python
## Визуализация: точки обучающей выборки и решающая граница (только по двум признакам)

#Нарисуем, как модель, обученная на длине и ширине лепестка, проводит границу между классами.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# 1. Загрузка и фильтрация Iris (только классы 1 и 2)
iris = load_iris()
X = iris.data[iris.target != 0]           # оставляем Versicolor и Virginica
y = (iris.target[iris.target != 0] - 1)    # перекодируем в 0 и 1

# 2. Выбираем только два признака: petal length (idx=2) и petal width (idx=3)
X2 = X[:, [2, 3]]

# 3. Делим на train/test
X2_train, X2_test, y_train, y_test = train_test_split(
    X2, y, test_size=0.2, random_state=42, stratify=y
)

# 4. Обучаем логистическую регрессию градиентным спуском на этих двух признаках
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# инициализация
w = np.zeros(2)
b = 0.0
lr = 0.01
epochs = 10000
m = len(y_train)

for _ in range(epochs):
    z = X2_train.dot(w) + b
    a = sigmoid(z)
    dw = (1/m) * X2_train.T.dot(a - y_train)
    db = (1/m) * np.sum(a - y_train)
    w -= lr * dw
    b -= lr * db

# 5. График
plt.figure(figsize=(8,6))

# 5.1 Точки train (синие и красные)
for cls, color, marker in zip([0,1], ['blue','red'], ['o','s']):
    mask = (y_train == cls)
    plt.scatter(
        X2_train[mask, 0], X2_train[mask, 1],
        c=color, marker=marker, edgecolor='k',
        label=f'Класс {cls} (train)'
    )

# 5.2 Решaющая граница: w1*x1 + w2*x2 + b = 0
x_vals = np.linspace(X2_train[:,0].min()-0.2, X2_train[:,0].max()+0.2, 200)
y_vals = -(b + w[0] * x_vals) / w[1]
plt.plot(x_vals, y_vals, 'k--', label='Граница σ=0.5')

plt.xlabel('petal length (cm)')
plt.ylabel('petal width  (cm)')
plt.title('Train точки и решающая граница')
plt.legend()
plt.grid(True)
plt.show()
```
