```python
## :       (   )

#,  ,      ,    .

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# 1.    Iris (  1  2)
iris = load_iris()
X = iris.data[iris.target != 0]           #  Versicolor  Virginica
y = (iris.target[iris.target != 0] - 1)    #   0  1

# 2.    : petal length (idx=2)  petal width (idx=3)
X2 = X[:, [2, 3]]

# 3.   train/test
X2_train, X2_test, y_train, y_test = train_test_split(
    X2, y, test_size=0.2, random_state=42, stratify=y
)

# 4.         
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

# 
w = np.zeros(2)
b = 0.0
lr = 0.01
epochs = 10000
m = len(y_train)

for _ in range(epochs):
    z = X2_train.dot(w) + b
    a = sigmoid(z)
    dw = (1/m) * X2_train.T.dot(a - y_train)
    db = (1/m) * np.sum(a - y_train)
    w -= lr * dw
    b -= lr * db

# 5. 
plt.figure(figsize=(8,6))

# 5.1  train (  )
for cls, color, marker in zip([0,1], ['blue','red'], ['o','s']):
    mask = (y_train == cls)
    plt.scatter(
        X2_train[mask, 0], X2_train[mask, 1],
        c=color, marker=marker, edgecolor='k',
        label=f' {cls} (train)'
    )

# 5.2 a : w1*x1 + w2*x2 + b = 0
x_vals = np.linspace(X2_train[:,0].min()-0.2, X2_train[:,0].max()+0.2, 200)
y_vals = -(b + w[0] * x_vals) / w[1]
plt.plot(x_vals, y_vals, 'k--', label=' =0.5')

plt.xlabel('petal length (cm)')
plt.ylabel('petal width  (cm)')
plt.title('Train    ')
plt.legend()
plt.grid(True)
plt.show()
```
